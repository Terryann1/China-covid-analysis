DATA VISUALIZATION

```python
#SCATTER PLOT
x=covid['Total Cases'].iloc[:20]
y=covid['Total Death'].iloc[:20]
plt.rcParams['figure.figsize']=(20,10)
xlabel='Total cases'
ylabel='Total death'
plt.title('scatter plot')
plt.scatter(x,y)

```

```python
#Line chart
x=covid['Total Cases'].iloc[:20]
y=covid['Total Death'].iloc[:20]
plt.rcParams['figure.figsize']=(20,10)
xlabel='Total cases'
ylabel='Total death'
plt.title('line plot')
plt.plot(x,y)
```

```python
#SCATTER PLOT AND LINE CHART
x=covid['Total Cases'].iloc[:20]
y=covid['Total Death'].iloc[:20]
plt.rcParams['figure.figsize']=(20,10)
plt.scatter(x,y)
plt.title('Line and scatter plot')
plt.plot(x,y)
```

```python
#BAR CHART
x=covid['Total Cases'].iloc[:20]
plt.rcParams['figure.figsize']=(20,10)
plt.title('Bar graph')
sns.countplot(x)
```

```python
#BAR CHART
x=covid['Total Death'].iloc[:20]
plt.rcParams['figure.figsize']=(20,10)
plt.title('Bar graph')
sns.countplot(x)
```
